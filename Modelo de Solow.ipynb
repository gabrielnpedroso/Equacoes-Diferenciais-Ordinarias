```python
import numpy as np
import matplotlib.pyplot as plt

# Variáveis Macro Econômicas - Parâmetros do modelo

s = 0.3  # Taxa de poupança

alpha = 0.3  # Elasticidade da produção em relação ao capital

delta = 0.05  # Taxa de depreciação

n = 0.02  # Taxa de crescimento populacional

# Função que define a EDO
def solow_model(k, t):
    return s * k**alpha - (delta + n) * k

# Método de Euler para resolver a EDO
def euler_method(func, k0, t, dt):
    result = [k0]
    for i in range(1, len(t)):
        k0 += func(k0, t[i-1]) * dt
        result.append(k0)
    return result

# Intervalos, passos e valor inicial
t = np.linspace(0, 100, 1000)  # Intervalo
k0 = 1.0  # Valor inicial
dt = t[1] - t[0]  # Passos

# Resolvendo numericamente a EDO
k_values = euler_method(solow_model, k0, t, dt)

# Plotando os resultados
plt.plot(t, k_values)
plt.xlabel('Tempo')
plt.ylabel('Estoque de Capital por Trabalhador (k)')
plt.title('Modelo de Crescimento de Solow')
plt.show()
```
